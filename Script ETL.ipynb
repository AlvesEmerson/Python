```python
import pandas as pd
import csv
import pymysql
import MySQLdb
from pandasql import sqldf
from datetime import datetime
import locale
```

```python
usuarios = pd.read_csv('Usuarios.csv')
produtos = pd.read_csv('Produtos.csv')
```

```python
print(usuarios)
```

```python
usuarios = usuarios.rename_axis('id_usuario').reset_index().rename(columns=
                           {'id_usuario': 'id_usuario'
                               , 'Nome': 'nome'
                               , 'Email': 'email'
                               , 'Estado': 'estado'})
usuarios
```

```python
usuarios.dtypes
```

```python
print(produtos)
```

```python
produtos = produtos.rename_axis('id_produto').reset_index().rename(columns=
                           {'id_produto': 'id_produto'
                               , 'Produto': 'nome'
                               , 'Valor': 'valor'
                               , 'Categoria': 'categoria'})

def corrigir_categoria(categoria):
    categoria = categoria.replace('Informática', 'Informatica').replace('Cerulares', 'Celulares').replace('Ganes', 'Games')
    return categoria

produtos['categoria'] = produtos['categoria'].apply(corrigir_categoria)
produtos
```

```python
produtos.dtypes
```

```python
conexao = pymysql.connect(db='comercial', user='root', passwd='12345678')
cursor = conexao.cursor()
print('Conexão bem sucedida!')

for indice, linha in usuarios.iterrows():
    cursor.execute("INSERT INTO usuario (id_usuario, nome, email, estado) values (%s, %s, %s, %s)",
                   (linha.id_usuario, linha.nome, linha.email, linha.estado))
    
for indice, linha in produtos.iterrows():
    cursor.execute("INSERT INTO produto (id_produto, nome, valor, categoria) values (%s, %s, %s, %s)",
                   (linha.id_produto, linha.nome, linha.valor, linha.categoria))

conexao.commit()
cursor.close()
```

```python
vendas = pd.read_csv('Venda.csv')
vendas
```

```python
vendas = vendas.rename_axis('id_venda').reset_index().drop(
    columns=['Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7'])
vendas
```

```python
vendas_final = sqldf('select * from vendas inner join usuarios on usuarios.Email = vendas.EmailUsuario')
vendas_final
```

```python
vendas_final2 = sqldf('select * from vendas_final inner join produtos on vendas_final.Produto = produtos.nome')
vendas_final2
```

```python
vendas_final3 = sqldf('select id_venda, DataVenda, Status, id_usuario, id_produto from vendas_final2')
vendas_final3
```

```python
vendas_final3 = vendas_final3.rename(columns=
                           {'id_venda': 'id_venda'
                               , 'DataVenda': 'data_venda'
                               , 'Status': 'status'
                               , 'id_usuario': 'id_usuario'
                               , 'id_produto': 'id_produto'})
vendas_final3
```

```python
vendas_final3.dtypes
```

```python
#vendas_final3['data_venda'] = pd.to_datetime(vendas_final3['data_venda'], format = '%d/%m/%Y')
```

```python
locale.setlocale(locale.LC_ALL, 'pt_BR')
formato = "%m/%d/%Y"
vendas_final3.data_venda = vendas_final3.data_venda.apply(lambda linha: datetime.strptime(linha, formato))
```

```python
vendas_final3.dtypes
```

```python
conexao = pymysql.connect(db='comercial', user='root', passwd='12345678')
cursor = conexao.cursor()
print('Conexão bem sucedida!')

for indice, linha in vendas_final3.iterrows():
    cursor.execute("INSERT INTO venda (id_venda, data_venda, status, id_usuario, id_produto) values (%s, %s, %s, %s, %s)",
                   (linha.id_venda, linha.data_venda, linha.status, linha.id_usuario, linha.id_produto))

conexao.commit()
cursor.close()
```
